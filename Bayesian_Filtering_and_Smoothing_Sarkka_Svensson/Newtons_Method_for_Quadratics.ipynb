Derivation of the Ordinary Least Squares solution to Linear Regression from Newton's Method

```julia
# Define initial parameters
w₀ = [5.0, -3.5]
noise_std = 10
noise_dist = Normal(0, noise_std)

# Define X-values and generate sample data
x = 0:0.1:50
y = w₀[1] .+ x.*w₀[2] .+ rand(noise_dist, length(x))

#= Solve the linear regression problem using Newton's method
The least squares error function = 0.5 * Σ(x*w - y)^2

X = Nx2
W = 2x1
Y = Nx1

g = X'(XW-y)
g = X'Xw - X'y

The Hessian H is thus 
H = X'X

Thus, Newton's update Wₖ₊₁ = Wₖ - H⁻¹g is
wₖ₊₁ = wₖ - (X'X)⁻¹(X'Xwₖ - X'y)
wₖ₊₁ = wₖ - (X'X)⁻¹(X'X)wₖ + (X'X)⁻¹(X'y)
wₖ₊₁ = wₖ - wₖ + (X'X)⁻¹(X'y)
wₖ = (X'X)⁻¹(X'y)

Since error function is quadratic, Newton's method converges in one step
w = (X'X)⁻¹(X'y)

This is just the OLS solution haha
=#

# Solve linear regression problem
X = hcat(ones(length(x)), x)
y = y
w = (X'*X) \ (X'*y)

y_est = X*w

# Plot dataset
scatter(x,y, label="Data", xlabel="x", ylabel="y")
plot!(x, y_est, label="Best Fit", linewidth=4)
title!("Newton's Method Regression")
```
