```julia
using Distributions
using Plots
```

Gaussian random follows the equations

xt = xt-1 + qk-1 where q ~ N(0, Q)

yt = xt +  rt where r ~ N(0, R)

```julia
using Plots

# Define the parameters of the walk
Q = 2.0  # Variance for the hidden state transitions
R = 1.0  # Variance for the emission model
num_steps = 100  # Number of steps for simulation

# Define the distributions
q_dist = Normal(0, Q)
r_dist = Normal(0, R)

# Initialize storage for hidden states and observations
x = zeros(num_steps)  # Latent state
y = zeros(num_steps)  # Observed state

for t in 2:num_steps
    x[t] = x[t-1] + rand(q_dist)  # Update the latent state
    y[t] = x[t] + rand(r_dist)    # Generate the observation
end

# Set the figure size and adjust the margins
plot(x, label="Latent State x", size=(800, 400), left_margin=10Plots.mm, bottom_margin=10Plots.mm)
plot!(y, label="Observed y", linestyle=:dash)

title!("Gaussian Random Walk")
xlabel!("Time")
ylabel!("Sample AU")
```

Define a Poisson Random Walk

xt = xt-1 + qk-1 where q ~ Poisson(λ)

yt = xt +  rt where r ~ N(0, R)

```julia
# Define parameters
λ = 2.0          # Poisson rate (controls the average step size)
R = 3.0        # Variance for the emission model
num_steps = 100  # Number of steps for simulation

# Define the distributions
poisson_dist = Poisson(λ)
r_dist = Normal(0, R)

# Initialize storage for hidden states and observations
x = zeros(Int, num_steps)  # Latent state (integer)
y = zeros(num_steps)       # Observed state

# Generate the Poisson random walk
for t in 2:num_steps
    x[t] = x[t-1] + rand(poisson_dist)   # Update latent state with Poisson step
    y[t] = x[t] + rand(r_dist)           # Generate observation with Gaussian noise
end

# Plot the results
plot(x, label="Latent State x (Poisson Walk)", size=(800, 400))
plot!(y, label="Observed y", linestyle=:dash)

title!("Poisson Random Walk")
xlabel!("Time")
ylabel!("Sample AU")
```
